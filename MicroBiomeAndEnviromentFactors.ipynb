```python
from sklearn.model_selection import train_test_split
from sklearn import linear_model
from sklearn import svm
import matplotlib.pyplot as plt
import random as rd
import numpy as np
rd.seed(0)

def loadClassifier(CLS):
    classifier="SVM"
    if CLS=="LASSO":
        title="LASSO"
        classifier=linear_model.LogisticRegression(penalty='l2')
    elif CLS=="SVM":
        title="Support Vector Machine"
        classifier= svm.SVC(kernel='linear', probability=True)
    elif CLS=="NB":
        title="Naive Bayes"
        classifier =GaussianNB()
    elif(CLS=='RF'):
        title="Random Forest"
        classifier = RandomForestClassifier(n_estimators=int(100))
    else:
        title="Support Vector Machine"
        classifier= svm.SVC(kernel='linear', probability=True)
        print("Unsuppored keyword for specified classifier, using default linear SVM")
    return classifier,title


def CFV(trainData,labels,classifierName,nfolds,testSize=0.2,vis=True):
    
    accuList=[]
    for i in range(0,nfolds):
        X_train, X_test, y_train, y_test = train_test_split(trainData, labels, test_size=0.25,random_state= rd.randint(0,100))
        clf,title=loadClassifier(classifierName)
        if clf in ["LASSO"]:
            clf.fit(X_train)
        else:
            clf.fit(X_train,y_train) 
        y_predict=clf.predict(X_test)
        print(y_predict)
        accu=getAccuracies(y_test,y_predict)
        accuList.append(accu)
#    print("Accuracy Across Folds",accuList)
    print ("Mean Accuracy Across Folds",np.mean(accuList))
    if vis:
    
        
        plt.figure()
        plt.title("Prediction Accuracies in Each Fold "+ "(" + title + ")")
        plt.axhline(y=np.mean(accuList), color='r', linestyle='-',label="Mean Accurracy Across Folds")
        plt.plot(range(1,nfolds+1),accuList)
        plt.xlabel("Fold")
        plt.legend()
        plt.ylabel("Accuracy")
        plt.show()

def getAccuracies(labels,predictions):
    T=0
    F=0
    for i in range(len(labels)):
        if labels[i]==predictions[i]:
            T+=1
        else:
            F+=1
    return(T/float(len(labels)))

def LeaveOneOut(matrix,labels,classfierName,vis=True):
    predictionLabels=[]
    clf,title=loadClassifier(classfierName)
    for i in range(len(matrix)):
        trainMatrix=np.delete(matrix,i,0)
        trainLabels=np.delete(labels,i,0)
        clf.fit(trainMatrix,trainLabels)
        testMatrix=[]
        testMatrix.append(matrix[i])
        predictionLabels.append(clf.predict(testMatrix)[0])
        
    return predictionLabels

```

```python
def perf_measure(y_actual, y_hat):
    TP = 0
    FP = 0
    TN = 0
    FN = 0

    for i in range(len(y_hat)): 
        if y_actual[i]==y_hat[i]==1:
           TP += 1
        if y_hat[i]==1 and y_actual[i]!=y_hat[i]:
           FP += 1
        if y_actual[i]==y_hat[i]==0:
           TN += 1
        if y_hat[i]==0 and y_actual[i]!=y_hat[i]:
           FN += 1

    return(TP, FP, TN, FN)


```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns

def make_confusion_matrix(cf,
                          group_names=None,
                          categories='auto',
                          count=True,
                          percent=True,
                          cbar=True,
                          xyticks=True,
                          xyplotlabels=True,
                          sum_stats=True,
                          figsize=None,
                          cmap='BYlGn',
                          title=None):


    # CODE TO GENERATE TEXT INSIDE EACH SQUARE
    blanks = ['' for i in range(cf.size)]

    if group_names and len(group_names)==cf.size:
        group_labels = ["{}\n".format(value) for value in group_names]
    else:
        group_labels = blanks

    if count:
        group_counts = ["{0:0.0f}\n".format(value) for value in cf.flatten()]
    else:
        group_counts = blanks

    if percent:
        group_percentages = ["{0:.2%}".format(value) for value in cf.flatten()/np.sum(cf)]
    else:
        group_percentages = blanks

    box_labels = [f"{v1}{v2}{v3}".strip() for v1, v2, v3 in zip(group_labels,group_counts,group_percentages)]
    box_labels = np.asarray(box_labels).reshape(cf.shape[0],cf.shape[1])


    # CODE TO GENERATE SUMMARY STATISTICS & TEXT FOR SUMMARY STATS
    if sum_stats:
        #Accuracy is sum of diagonal divided by total observations
        accuracy  = np.trace(cf) / float(np.sum(cf))

        #if it is a binary confusion matrix, show some more stats
        if len(cf)==2:
            #Metrics for Binary Confusion Matrices
            precision = cf[1,1] / sum(cf[:,1])
            recall    = cf[1,1] / sum(cf[1,:])
            f1_score  = 2*precision*recall / (precision + recall)
            stats_text = "\n\nAccuracy={:0.3f}\nPrecision={:0.3f}\nRecall={:0.3f}\nF1 Score={:0.3f}".format(
                accuracy,precision,recall,f1_score)
        else:
            stats_text = "\n\nAccuracy={:0.3f}".format(accuracy)
    else:
        stats_text = ""


    # SET FIGURE PARAMETERS ACCORDING TO OTHER ARGUMENTS
    if figsize==None:
        #Get default figure size if not set
        figsize = plt.rcParams.get('figure.figsize')

    if xyticks==False:
        #Do not show categories if xyticks is False
        categories=False


    # MAKE THE HEATMAP VISUALIZATION
    plt.figure(figsize=figsize)
    sns.heatmap(cf,annot=box_labels,fmt="",cmap=cmap,cbar=cbar,xticklabels=categories,yticklabels=categories)
    

    if xyplotlabels:
        plt.ylabel('True label')
        plt.xlabel('Predicted label' + stats_text)
    else:
        plt.xlabel(stats_text)
    
    if title:
        plt.title(title)
    plt.show()
    return accuracy,precision,recall,f1_score

def binaryLabelVis(y,y_pred,categories,title="title",cmap="OrRd"):    
    cf_matrix = confusion_matrix(y, y_pred)
    labels = ["True Neg","False Pos","False Neg","True Pos"]
    accuracy,precision,recall,f1_score=make_confusion_matrix(cf_matrix, 
                          group_names=labels,
                          categories=categories, 
                          cmap=cmap,title=title) 
    return accuracy,precision,recall,f1_score     
```

Let's look at all three enviroment factors together and the predicted results by different number of selected features:

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2,f_classif
from sklearn.metrics import precision_score,recall_score, confusion_matrix, classification_report, accuracy_score, f1_score


def plotFeatureTrend(matrix,labels,minFeaure=1,maxFeature=101,title="",increment=1,scoringFunction=chi2):
    kList=[]
    
    cur=minFeaure
    while cur<=maxFeature:
        kList.append(cur)
        cur+=increment
    
    accuList=[]
    precisionList=[]
    for myK in kList:
        FS=SelectKBest(scoringFunction,k=myK)
#        FS= VarianceThreshold()
        
        selectedMatrix=FS.fit_transform(matrix,labels)
#        print(feature_index)
#        print(rowHeader[feature_index]) 
        predictions=LeaveOneOut(selectedMatrix,labels,"SVM")
        accu=getAccuracies(labels,predictions)
        f1score=f1_score(labels, predictions)
        accuList.append(accu)
        precisionList.append(f1score)
    fig=plt.figure(dpi=80)
    plt.ylim([0,1])
    plt.title(title)
    plt1=plt.plot(kList,accuList)
    plt.xlabel("Number of Features Selected")
    plt.ylabel("F1 Score")
    ax2=plt.twinx()
    plt2=plt.plot(kList,precisionList,c="red")
    ax2.set_ylabel("F1 Score")
    ax2.set_ylim([0,1])
    
    plt.legend((plt1[0],plt2[0]),("Accurary","F1 Score"))
    plt.show()
    return kList,accuList
```

```python
def predictWithFS(cat,matrix,label,topFeature=30,CLS="SVM",scoringFunction=chi2,title=""):
    FS = SelectKBest(scoringFunction, k=topFeature)
    fitResult=FS.fit(matrix,label)
    selectedFeatureIndex=2
    scores=(fitResult.scores_)
    selectedMatrix = FS.fit_transform(matrix, label)
    predictions = LeaveOneOut(selectedMatrix, label, CLS)
    accuracy,precision,recall,f1_score=binaryLabelVis(label, predictions, cat, title=title,cmap="BuGn")
    return accuracy,precision,recall,f1_score,scores,selectedFeatureIndex
```
