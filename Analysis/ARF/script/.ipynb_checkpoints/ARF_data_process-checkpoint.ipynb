```python
def CombineData(rank):

    # load dataset
    df_sp = pd.read_csv('../data/ARF_16S_ctb_SP_' + rank + '.csv',index_col = 0)
    df_sp.columns = df_sp.columns.str.replace('ctb_'+rank.lower()+'.', '', regex=False)
    
    df_win = pd.read_csv('../data/ARF_16S_ctb_WIN_' + rank + '.csv',index_col = 0)
    df_win.columns = df_win.columns.str.replace('ctb_'+rank.lower()+'.', '', regex=False)

    df_combine = pd.concat([df_sp, df_win], axis=0, join="inner")
    return df_combine 
```

```python
### combine family data 
cts_family_combine = CombineData('Family')
```
